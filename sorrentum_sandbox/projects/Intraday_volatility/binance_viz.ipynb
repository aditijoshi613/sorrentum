```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#importing the binance data
binance_data = pd.read_csv('clean_binance.csv')
#binance_data = binance_data.drop(['Unnamed: 0'],axis = 1) # just in case the index issue arises again
binance_data['GK_model'] = .5 * (np.power( np.log(binance_data['high']/binance_data['low']) ,2) ) \
                            - (2*np.log(2) - 1) * np.power(np.log(binance_data['close']/binance_data['open']),2)
```

```python
binance_data = binance_data[:10000].pivot(index='timestamp', columns='currency_pair', values='GK_model')
binance_data.plot(figsize=(10, 10))
```
