```python
import pandas as pd
import pytz
```

```python
OUTLIER_TRADES_PATH = "/Users/firstsingularity/Desktop/top_100_outliers_window_5760_sig_5.csv"
RP_DATA_PATH = "/Users/firstsingularity/Desktop/rp_data_top10_crypto_2021.csv"
```

```python
outliers_df = pd.read_csv(OUTLIER_TRADES_PATH).query('"2021-01-01 00:00:00" <= timestamp < "2022-01-01 00:00:00"')
outliers_df['abs_norm_ret'] = abs(outliers_df['normalized_ret'])
outliers_df = outliers_df.sort_values(by=['abs_norm_ret'])
outliers_df.head()
```

```python
rp_df = pd.read_csv(RP_DATA_PATH)[["TIMESTAMP_UTC", "ENTITY_NAME", "ENTITY_DETECTION_TYPE", "ENTITY_RELEVANCE", "ENTITY_SENTIMENT", "ENTITY_SENTIMENT_CONFIDENCE", "ENTITY_TEXT_SENTIMENT", "ENTITY_TEXT_SENT_CONFIDENCE", "CATEGORY", "FACT_LEVEL", "DOCUMENT_TYPE", "TITLE", "SOURCE_NAME", "SOURCE_RANK"]].rename(columns={"TIMESTAMP_UTC": "timestamp"})
rp_df.timestamp = pd.DatetimeIndex(rp_df["timestamp"]).tz_localize(pytz.utc).tz_convert(pytz.timezone("America/New_York"))
rp_df = rp_df.set_index("timestamp", drop=False).loc['2021-01-01':'2022-01-01'].sort_index().drop(columns=['timestamp'])
rp_df.head()
```

```python
range_start = pd.Timedelta(minutes=1)
range_end = pd.Timedelta(minutes=1)
```

```python
query_results = {}

for index, row in outliers_df.iterrows():
    center_datetime = pd.to_datetime(row["timestamp"], infer_datetime_format=True)
    query_result = rp_df[(rp_df.index >= center_datetime - range_start) & (rp_df.index <= center_datetime + range_end)]
    query_results["{0}-({1})".format(row["currency_pair"], row["timestamp"])] = query_result
```

```python
for key in query_results.keys():
    print(key)
```
